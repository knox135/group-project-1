```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
```

```python

sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
url = "https://api-nba-v1.p.rapidapi.com/players"
api_key_str = '?key='+sports_data_api_key
querystring = {"team":"24","season":"2023"}

headers = {
	"X-RapidAPI-Key": sports_data_api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
response_knicks = response.json()
print(json.dumps(response_knicks, indent=4))
```

```python
knicks_df = pd.json_normalize(response_knicks['response']).reset_index(drop=True)
knicks_df
```

```python
knicks_df_csv = knicks_df.to_csv('Resources/New York Knicks.csv', index=False)
```

```python
import time

data_frames = []

def get_player_stats(player_id):
    time.sleep(30)
    return pd.json_normalize(requests.get(url+"/statistics", headers=headers, params={"id":player_id, "season":"2023"}).json()['response'])
    
```

```python
for id in knicks_df['id']:
    df = get_player_stats(id)
    data_frames.append(df)
```

```python
combined_df = pd.concat(data_frames, ignore_index=True)
combined_df
```

```python
knicks_stats = combined_df.to_csv('Resources/Stats New York Knicks.csv', index=False)
```
