```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
```

```python
url = "https://api-nba-v1.p.rapidapi.com/standings"
api_key_str = '?key='+sports_data_api_key
```

```python
seasons = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
# no data for 2015, 16, 17
querystring21 = {"league":"standard","season":"2021"}

headers = {
	"X-RapidAPI-Key": "57ecd31499mshfd9806745498469p111422jsnd51630f333ce",
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response21 = requests.get(url, headers=headers, params=querystring21)
response21_json = response21.json()

print(json.dumps(response21_json['response'], indent=4))
```

```python
standings21_df = pd.json_normalize(response21_json['response']).reset_index(drop=True)
standings21_df.sort_values(by='gamesBehind', ascending=True)
```

```python
standings21_df_csv = standings_df.to_csv('Resources/2021 Standings.csv', index=False)
standings21_df_csv
```
