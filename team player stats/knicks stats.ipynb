```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
```

```python
url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

querystring = {"id":"978","season":"2023"}

headers = {
	"X-RapidAPI-Key": sports_data_api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response_hartenstein = requests.get(url, headers=headers, params=querystring).json()
hartenstein_df = pd.json_normalize(response_hartenstein['response']).reset_index(drop=True)
hartenstein_df = hartenstein_df[['player.firstname','player.lastname','pos','points','min','fgm','fga','fgp','ftm','fta','tpm','tpa', 'tpp', 'offReb', 'defReb', 'totReb', 'assists', 'pFouls','steals', 'turnovers', 'blocks', 'plusMinus',]]
hartenstein_df
```
