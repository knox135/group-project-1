```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
```

```python


url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

querystring = {"id":"519","season":"2023"}

headers = {
	"X-RapidAPI-Key": sports_data_api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response_kat = requests.get(url, headers=headers, params=querystring).json()

print(json.dumps(response_kat, indent=4))
```

```python
response_kat_df = pd.json_normalize(response_kat['response']).reset_index(drop=True)
response_kat_df
```

```python
response_kat_df_csv = response_kat_df.to_csv('Resources/timberwolves stats/KAT.csv', index=False)
```
