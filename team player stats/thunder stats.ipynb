```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
```

```python
url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

querystring = {"id":"972","season":"2023"}

headers = {
	"X-RapidAPI-Key": sports_data_api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response_alexander = requests.get(url, headers=headers, params=querystring).json()
#print(json.dumps(response_alexander, indent=4))
alexander_df = pd.json_normalize(response_alexander['response']).reset_index(drop=True)
alexander_df = alexander_df[['player.firstname','player.lastname','pos','points','min','fgm','fga','fgp','ftm','fta','tpm','tpa', 'tpp', 'offReb', 'defReb', 'totReb', 'assists', 'pFouls','steals', 'turnovers', 'blocks', 'plusMinus',]]
alexander_df
```
