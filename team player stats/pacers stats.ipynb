```python
import sys
from dotenv import load_dotenv
import os
import requests
import pandas as pd
import json
```

```python
load_dotenv()
load_dotenv('.env')
sports_data_api_key = os.getenv('SPORTS_DATA_API_KEY')
```

```python
url = "https://api-nba-v1.p.rapidapi.com/players/statistics"

querystring = {"id":"522","season":"2023"}

headers = {
	"X-RapidAPI-Key": sports_data_api_key,
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}

response_turner = requests.get(url, headers=headers, params=querystring).json()
turner_df = pd.json_normalize(response_turner['response']).reset_index(drop=True)
turner_df = turner_df[['player.firstname','player.lastname','pos','points','min','fgm','fga','fgp','ftm','fta','tpm','tpa', 'tpp', 'offReb', 'defReb', 'totReb', 'assists', 'pFouls','steals', 'turnovers', 'blocks', 'plusMinus',]]
turner_df
```
