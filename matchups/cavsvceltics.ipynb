```python
#import requirements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from prophet import Prophet
```

```python
#Read in Caveliers player csv data
Mobley_df = pd.read_csv('./resources/caveliers stats/Mobley.csv')
Okoro_df = pd.read_csv('./resources/caveliers stats/Okoro.csv')
Garland_df = pd.read_csv('./resources/caveliers stats/Garland.csv')
LeVert_df = pd.read_csv('./resources/caveliers stats/Levert.csv')
Strus_df = pd.read_csv('./resources/caveliers stats/Strus.csv')
Wade_df = pd.read_csv('./resources/caveliers stats/Wade.csv')
Merrill_df = pd.read_csv('./resources/caveliers stats/Merrill.csv')
#Read in Celtics player csv files
tatum_df = pd.read_csv('./resources/celtics stats/tatum.csv')
horford_df = pd.read_csv('./resources/celtics stats/horford.csv')
holiday_df = pd.read_csv('./resources/celtics stats/holiday.csv')
white_df = pd.read_csv('./resources/celtics stats/white.csv')
brown_df = pd.read_csv('./resources/celtics stats/brown.csv')
hauser_df = pd.read_csv('./resources/celtics stats/hauser.csv')
kornet_df = pd.read_csv('./resources/celtics stats/kornet.csv')
pritchard_df = pd.read_csv('./resources/celtics stats/pritchard.csv')
#Read in schedule for dates
cavs_schedule = pd.read_csv('./resources/schedules/cavs schedule.csv',parse_dates=True)
celtics_schedule = pd.read_csv('./resources/schedules/celtics Schedule.csv',parse_dates=True)
```

```python
#set date format type for schedules
cavs_schedule['date'] = pd.to_datetime(cavs_schedule['date']).dt.tz_localize(None)
celtics_schedule['date'] = pd.to_datetime(celtics_schedule['date']).dt.tz_localize(None)
```

```python
#Combine Caveliers player data with dates of games
Mobley_df = pd.concat([cavs_schedule, Mobley_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Okoro_df = pd.concat([cavs_schedule, Okoro_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Garland_df = pd.concat([cavs_schedule, Garland_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
LeVert_df = pd.concat([cavs_schedule, LeVert_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Strus_df = pd.concat([cavs_schedule, Strus_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Wade_df = pd.concat([cavs_schedule, Wade_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Merrill_df = pd.concat([cavs_schedule, Merrill_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
#Combine Celtics player data with dates of games
tatum_df = pd.concat([celtics_schedule, tatum_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
horford_df = pd.concat([celtics_schedule, horford_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
holiday_df = pd.concat([celtics_schedule, holiday_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
white_df = pd.concat([celtics_schedule, white_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
brown_df = pd.concat([celtics_schedule, brown_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
hauser_df = pd.concat([celtics_schedule, hauser_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
kornet_df = pd.concat([celtics_schedule, kornet_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
pritchard_df = pd.concat([celtics_schedule, pritchard_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
```

```python
#celtics trends
tatum_trend = tatum_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Tatum')
horford_trend = horford_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Horford')
holiday_trend = holiday_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Holiday')
white_trend = white_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('White')
brown_trend = brown_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Brown')
hauser_trend = hauser_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Hauser')
kornet_trend = kornet_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Kornet')
pritchard_trend = pritchard_trend[['player.lastname','trend','yhat',]].reset_index().rename(columns={'index':'game'}).fillna('Pritchard')
#cavs trends
Mobley_trend = Mobley_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Mobley')
Okoro_trend = Okoro_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Okoro')
Garland_trend = Garland_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Garland')
Strus_trend = Strus_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Strus')
Wade_trend = Wade_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Wade')
Merrill_trend = Merrill_trend[['player.lastname','trend','yhat']].reset_index().rename(columns={'index':'game'}).fillna('Merrill')

```

```python
#team comparison
cavs_trend = pd.concat([Mobley_trend,Okoro_trend,Garland_trend,Strus_trend,Wade_trend,Merrill_trend],axis=0).rename(columns={'player.lastname':'name'})
celtics_trend = pd.concat([tatum_trend,horford_trend,holiday_trend,white_trend,brown_trend,hauser_trend,kornet_trend,pritchard_trend],axis=0).rename(columns={'player.lastname':'name'})
cavs_compare = cavs_trend[['game','yhat']].groupby(['game']).mean()
celtics_compare = celtics_trend[['game','yhat']].groupby(['game']).mean()
display(cavs_compare.tail(7))
display(celtics_compare.tail(7))
```
