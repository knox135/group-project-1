```python
#import requirements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from prophet import Prophet
```

```python
#read in knicks player stats
hartenstein_df = pd.read_csv('../resources/knicks stats/hartenstein.csv')
Anunoby_df = pd.read_csv('../resources/knicks stats/Anunoby.csv')
Brunson_df = pd.read_csv('../resources/knicks stats/Brunson.csv')
McBride_df = pd.read_csv('../resources/knicks stats/McBride.csv')
hart_df = pd.read_csv('../resources/knicks stats/hart.csv')
DiVincenzo_df = pd.read_csv('../resources/knicks stats/DiVincenzo.csv')
Arcidiacono_df = pd.read_csv('../resources/knicks stats/Arcidiacono.csv')
burks_df = pd.read_csv('../resources/knicks stats/burks.csv')
#read pacers player stats
turner_df = pd.read_csv('../resources/pacers stats/turner.csv')
Siakam_df = pd.read_csv('../resources/pacers stats/Siakam.csv')
Nesmith_df = pd.read_csv('../resources/pacers stats/Nesmith.csv')
Nembhard_df = pd.read_csv('../resources/pacers stats/Nembhard.csv')
Haliburton_df = pd.read_csv('../resources/pacers stats/Haliburton.csv')
Toppin_df = pd.read_csv('../resources/pacers stats/Toppin.csv')
McConnell_df = pd.read_csv('../resources/pacers stats/McConnell.csv')
Sheppard_df = pd.read_csv('../resources/pacers stats/Sheppard.csv')
knicks_schedule = pd.read_csv('../resources/schedules/knicks schedule.csv',parse_dates=True)
pacers_schedule = pd.read_csv('../resources/schedules/pacers schedule.csv',parse_dates=True)
```

```python
#set schedules to datetime format
knicks_schedule['date'] = pd.to_datetime(knicks_schedule['date']).dt.tz_localize(None)
pacers_schedule['date'] = pd.to_datetime(pacers_schedule['date']).dt.tz_localize(None)
```

```python
#combine knicks players and game dates
hartenstein_df = pd.concat([knicks_schedule, hartenstein_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Anunoby_df = pd.concat([knicks_schedule, Anunoby_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Brunson_df = pd.concat([knicks_schedule, Brunson_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
McBride_df = pd.concat([knicks_schedule, McBride_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
hart_df = pd.concat([knicks_schedule, hart_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
DiVincenzo_df = pd.concat([knicks_schedule, DiVincenzo_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Arcidiacono_df = pd.concat([knicks_schedule, Arcidiacono_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
burks_df = pd.concat([knicks_schedule, burks_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
#combine players and game dates for pacers
turner_df = pd.concat([pacers_schedule, turner_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Siakam_df = pd.concat([pacers_schedule, Siakam_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Nesmith_df = pd.concat([pacers_schedule, Nesmith_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Nembhard_df = pd.concat([pacers_schedule, Nembhard_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Haliburton_df = pd.concat([pacers_schedule, Haliburton_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Toppin_df = pd.concat([pacers_schedule, Toppin_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
McConnell_df = pd.concat([pacers_schedule, McConnell_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
Sheppard_df = pd.concat([pacers_schedule, Sheppard_df], axis=1).fillna('0').drop([90,91,92,93,94,95,96,],axis=0)
```

```python
#use prophet to forecast knicks players
hartenstein_impact = hartenstein_df[['date','plusMinus']]
hartenstein_impact = hartenstein_impact.rename(columns={'date':'ds','plusMinus':'y'})
hartenstein_model = Prophet()
hartenstein_model.fit(hartenstein_impact)
hartenstein_future = hartenstein_model.make_future_dataframe(periods=7,freq='D')
hartenstein_forecast = hartenstein_model.predict(hartenstein_future)
hartenstein_forecast[['yhat','trend']].plot(legend='impact trend',title=' Impact forecast Trend')
hartenstein_trend = pd.concat([hartenstein_df['player.lastname'], hartenstein_forecast],axis=1)
display(hartenstein_trend[['player.lastname','yhat','trend']])

Anunoby_impact = Anunoby_df[['date','plusMinus']]
Anunoby_impact = Anunoby_impact.rename(columns={'date':'ds','plusMinus':'y'})
Anunoby_model = Prophet()
Anunoby_model.fit(Anunoby_impact)
Anunoby_future = Anunoby_model.make_future_dataframe(periods=7,freq='D')
Anunoby_forecast = Anunoby_model.predict(Anunoby_future)
Anunoby_forecast[['yhat','trend']].plot(legend='impact trend',title=' Impact forecast Trend')
Anunoby_trend = pd.concat([Anunoby_df['player.lastname'], Anunoby_forecast],axis=1)
display(Anunoby_trend[['player.lastname','yhat','trend']])

Brunson_impact = Brunson_df[['date','plusMinus']]
Brunson_impact = Brunson_impact.rename(columns={'date':'ds','plusMinus':'y'})
Brunson_model = Prophet()
Brunson_model.fit(Brunson_impact)
Brunson_future = Brunson_model.make_future_dataframe(periods=7,freq='D')
Brunson_forecast = Brunson_model.predict(Brunson_future)
Brunson_forecast[['yhat','trend']].plot(legend='impact trend',title=' Impact forecast Trend')
Brunson_trend = pd.concat([Brunson_df['player.lastname'], Brunson_forecast],axis=1)
display(Brunson_trend[['player.lastname','yhat','trend']])

McBride_impact = McBride_df[['date','plusMinus']]
McBride_impact = McBride_impact.rename(columns={'date':'ds','plusMinus':'y'})
McBride_model = Prophet()
McBride_model.fit(McBride_impact)
McBride_future = McBride_model.make_future_dataframe(periods=7,freq='D')
McBride_forecast = McBride_model.predict(McBride_future)
McBride_forecast[['yhat','trend']].plot(legend='impact trend',title=' Impact forecast Trend')
McBride_trend = pd.concat([McBride_df['player.lastname'], McBride_forecast],axis=1)
display(McBride_trend[['player.lastname','yhat','trend']])

hart_impact = hart_df[['date','plusMinus']]
hart_impact = hart_impact.rename(columns={'date':'ds','plusMinus':'y'})
hart_model = Prophet()
hart_model.fit(hart_impact)
hart_future = hart_model.make_future_dataframe(periods=7,freq='D')
hart_forecast = hart_model.predict(hart_future)
hart_forecast[['yhat','trend']].plot(legend='hart impact trend',title='hart Impact forecast Trend')
hart_trend = pd.concat([hart_df['player.lastname'], hart_forecast],axis=1)
display(hart_trend[['player.lastname','yhat','trend']])

DiVincenzo_impact = DiVincenzo_df[['date','plusMinus']]
DiVincenzo_impact = DiVincenzo_impact.rename(columns={'date':'ds','plusMinus':'y'})
DiVincenzo_model = Prophet()
DiVincenzo_model.fit(DiVincenzo_impact)
DiVincenzo_future = DiVincenzo_model.make_future_dataframe(periods=7,freq='D')
DiVincenzo_forecast = DiVincenzo_model.predict(DiVincenzo_future)
DiVincenzo_forecast[['yhat','trend']].plot(legend='DiVincenzo impact trend',title='DiVincenzo Impact forecast Trend')
DiVincenzo_trend = pd.concat([DiVincenzo_df['player.lastname'], DiVincenzo_forecast],axis=1)
display(DiVincenzo_trend[['player.lastname','yhat','trend']])

Arcidiacono_impact = Arcidiacono_df[['date','plusMinus']]
Arcidiacono_impact = Arcidiacono_impact.rename(columns={'date':'ds','plusMinus':'y'})
Arcidiacono_model = Prophet()
Arcidiacono_model.fit(Arcidiacono_impact)
Arcidiacono_future = Arcidiacono_model.make_future_dataframe(periods=7,freq='D')
Arcidiacono_forecast = Arcidiacono_model.predict(Arcidiacono_future)
Arcidiacono_forecast[['yhat','trend']].plot(legend='impact trend',title=' Impact forecast Trend')
Arcidiacono_trend = pd.concat([Arcidiacono_df['player.lastname'], Arcidiacono_forecast],axis=1)
display(Arcidiacono_trend[['player.lastname','yhat','trend']])

burks_impact = burks_df[['date','plusMinus']]
burks_impact = burks_impact.rename(columns={'date':'ds','plusMinus':'y'})
burks_model = Prophet()
burks_model.fit(burks_impact)
burks_future = burks_model.make_future_dataframe(periods=7,freq='D')
burks_forecast = burks_model.predict(burks_future)
burks_forecast[['yhat','trend']].plot(legend='burks impact trend',title='burks Impact forecast Trend')
burks_trend = pd.concat([burks_df['player.lastname'], burks_forecast],axis=1)
display(burks_trend[['player.lastname','yhat','trend']])

```
